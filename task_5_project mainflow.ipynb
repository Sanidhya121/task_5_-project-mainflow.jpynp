```python
questions=["1.How many people have heart disease and how many people doesn't have heart disease? ",
           "2.People of which sex has most heart disease  ? ",
           "3.People of which sex has which type of chest pain most ?"
           "4.People with which chest pain are most pron to have heart disease ?",
          ]
```

```python
#finding answer of first question 
#"1.How many people have heart disease and how many people doesn't have heart disease? "
# getting the values
df.target.value_counts()
```

```python
#ploting the bar chart of the first question
df.target.value_counts().plot(kind = 'bar',color=["orchid","salmon"])
plt.title("HEART DISEASE VALUES")
plt.xlabel("1 =Heart Disease, 0 = no heart Disease ")
plt.ylabel("Amount");
```

```python
#ploting pie chart
df.target.value_counts().plot(kind="pie",figsize = (8,6))
plt.legend(["Disease","No disease"]);
```

```python

#here we are checking the how many male and female in the dataset
df.sex.value_counts()
# 0 represent the female and 1 represent the male
# IN Sex column part : 0 represent "no disease" and 1 represent "disease" 
```

```python
#ploting a pie chart 
df.sex.value_counts().plot(kind='pie',figsize=(8,6))
plt.title('Male Female ratio')
plt.legend(['Male','Female']);
```

```python
#Question 2.People of which sex has most heart disease  ? 




pd.crosstab(df.target,df.sex)
```

```python
sns.countplot(x='target',data= df,hue = "sex")
plt.title("Heart disease frequency of  sex")
plt.xlabel("0 =No heart disease, 1= heart disease")
```

```python
#Question "3.People of which sex has which type of chest pain most ?"

#counting the values of different chest pain

df.cp.value_counts()
```
